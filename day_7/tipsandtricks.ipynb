# 03-How to rename columns

```python
df=pd.DataFrame({'A col':[1,2,3],'B col':[4,5,6]})
df.head()
```

```python
df.rename(columns={'A col':'col_a','B col':'b_col'},inplace=True)
df
```

```python
df.columns=['col_aa','col_bb']
df
```

```python
# repleace any charchter, string
df.columns=df.columns.str.replace('_','*')
df
```

```python
#  adding prefix
df=df.add_prefix('zesh_')
df
```

```python
#  adding prefix
df=df.add_suffix('zesh_')
df
```

```python
df.columns=['col_a','col_b']
df
```
